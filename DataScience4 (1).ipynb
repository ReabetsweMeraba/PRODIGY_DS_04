## Import necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load the dataset

```python
file_path = 'twitter_training.csv'  # Replace with the correct path if needed
data = pd.read_csv(file_path)
```

## Rename columns for clarity

```python
data.columns = ['ID', 'Topic', 'Sentiment', 'Text']
```

## Handle missing values by removing rows with null 'Text'

```python
cleaned_data = data.dropna(subset=['Text'])
```

## Analyze sentiment distribution

```python
sentiment_counts = cleaned_data['Sentiment'].value_counts()
```

## Plot the sentiment distribution

```python
plt.figure(figsize=(8, 5))
plt.bar(sentiment_counts.index, sentiment_counts.values, color=['blue', 'green', 'gray', 'red'])
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
